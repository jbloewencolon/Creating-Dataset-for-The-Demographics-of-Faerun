```python
def randomize_breakdown(breakdown):
    if not isinstance(breakdown, dict):
        breakdown = {
            "humans": 80.0,
            "halflings": 8.0,
            "dwarves": 5.0,
            "elves": 4.0,
            "misc.": 3.0
        }

    try:
        breakdown_values = [float(value.strip('%')) / 100.0 if isinstance(value, str) else value / 100.0 for value in breakdown.values()]
        breakdown_values = np.array(breakdown_values)
        breakdown_values = breakdown_values / breakdown_values.sum()
        random_breakdown_values = np.random.dirichlet(breakdown_values)
        random_breakdown = {k: round(v * 100, 1) for k, v in zip(breakdown.keys(), random_breakdown_values)}
        return random_breakdown
    except Exception as e:
        print(f"Error in randomizing breakdown: {e}")
        return None
```

```python
def get_known_and_unknown_populations(region_data):
    capital_data = region_data.get("capital", {})
    settlements_data = region_data.get("settlements", [])

    if isinstance(capital_data, dict):
        capital_population = capital_data.get("population")
        if capital_population is not None:
            capital_population = int(capital_population)
    else:
        capital_population = None

    known_populations = []
    unknown_city_count = 0

    if capital_population is not None:
        known_populations.append(capital_population)
    else:
        unknown_city_count += 1

    for settlement in settlements_data:
        if isinstance(settlement, dict):
            settlement_population = settlement.get("population")
            if settlement_population is not None:
                known_populations.append(int(settlement_population))
            else:
                unknown_city_count += 1

    return known_populations, unknown_city_count
```

```python
def create_dataframe(regions_and_nations):
    data_rows = []
    for region_data in regions_and_nations:
        region_name = region_data["region"]
        ruler = region_data.get("ruler", "")

        population_data = region_data.get("population", {})
        if isinstance(population_data, dict):
            population = population_data.get("total")
        else:
            population = population_data

        try:
            population = int(population)
        except (ValueError, TypeError):
            population = None

        # Initialize unknown_populations_list before the if condition
        known_populations, unknown_city_count = get_known_and_unknown_populations(region_data)
        total_known_population = sum(known_populations)
        remaining_population = max(0, (population or 0) - total_known_population)

        if unknown_city_count > 0 and remaining_population > 0:
            average_population = remaining_population // unknown_city_count
            populations_to_distribute = [average_population] * unknown_city_count
            populations_to_distribute[-1] += remaining_population - sum(populations_to_distribute)
            unknown_populations_list = populations_to_distribute

        if isinstance(population_data, dict):
            population = population_data.get("total")
            region_breakdown = population_data.get("breakdown", {})
        else:
            population = population_data
            region_breakdown = {}

        area = region_data.get("area", {}).get("miles_squared", "")
        military = region_data.get("military", "")
        languages = ", ".join([str(lang) for lang in region_data.get("languages", [])])
        religion = ", ".join([str(rel) for rel in region_data.get("religion", [])])
        economy = region_data.get("economy", {})
        exports = ", ".join([str(exp) for exp in (economy.get("exports") or [])])
        imports = ", ".join([str(imp) for imp in (economy.get("imports") or [])])
        sources = ", ".join([str(src) for src in region_data.get("sources", [])])

        settlements_data = region_data.get("settlements", [])  # Correct place for the definition
        capital_data = region_data.get("capital", {})

        if isinstance(capital_data, dict):
            capital_name = capital_data.get("name", "")
            capital_population = capital_data.get("population")
            if capital_population is not None:
                capital_population = int(capital_population)
            else:
                capital_population = 0
        else:
            capital_name = str(capital_data)  # This ensures capital_name is always defined, converting None to an empty string if necessary
            capital_population = 0

        capital_row = [capital_name, region_name, capital_name, capital_population, population, area, ruler, military, languages, religion, exports, imports, randomize_breakdown(region_breakdown), sources]
        data_rows.append(capital_row)

        for settlement_data in settlements_data:
            if isinstance(settlement_data, dict):
                settlement_name = settlement_data.get("name", "")
                settlement_population = settlement_data.get("population")
            else:
                settlement_name = settlement_data
                settlement_population = None

            if settlement_population is None:
                if unknown_populations_list:
                    settlement_population = unknown_populations_list.pop(0)
                else:
                    settlement_population = 0
            else:
                settlement_population = int(settlement_population)

            settlement_row = [settlement_name, region_name, capital_name, settlement_population, population, area, ruler, military, languages, religion, exports, imports, randomize_breakdown(region_breakdown), sources]
            data_rows.append(settlement_row)

    df = pd.DataFrame(data_rows, columns=["Settlement", "Region", "Capital", "Settlement_Population", "Region_Population", "Area", "Ruler", "Military", "Languages", "Religions", "Exports", "Imports", "Population_Breakdown", "Sources"])
    return df
```

```python
# Convert columns to numeric, coercing errors to NaN
df['settlement_population'] = pd.to_numeric(df['settlement_population'], errors='coerce')
df['region_population'] = pd.to_numeric(df['region_population'], errors='coerce')

# Define size scaling factors
size_scaling = {'Capital': 3, 'Town': 2, 'Settlement': 1}

# Add a 'scale_factor' column to the DataFrame
df['scale_factor'] = df['size'].map(size_scaling).fillna(1)

# Calculate the normalization of scale factors ahead of time
df['scale_factor_normalized'] = df.groupby('region')['scale_factor'].transform(lambda x: x / x.sum())

# Filter for settlements with missing or zero population
no_pop_settlements = df[df['settlement_population'].isna() | (df['settlement_population'] == 0)]

# Function to calculate distributed population for a region group
def distribute_population(group):
    # Use the first 'region_population' value, as all are assumed to be the same within the group
    total_region_pop = group['region_population'].iat[0]
    # Distribute population based on the pre-calculated normalized scale factor
    population_to_distribute = total_region_pop * group['scale_factor_normalized']
    # Assign the calculated population to the settlement_population column
    group['settlement_population'].update(population_to_distribute)
    return group

# Apply the function to each group of settlements within the same region
filled_pop_settlements = no_pop_settlements.groupby('region', group_keys=False).apply(distribute_population)
```

```python
def clean_column_list(column):
    cleaned_column = []
    for item in column:
        try:
            # Safely evaluate the string as a list
            evaluated_item = ast.literal_eval(item)
            if isinstance(evaluated_item, list):
                # Join the list into a string with comma-separated items
                cleaned_column.append(','.join(evaluated_item))
            else:
                # If the evaluated item is not a list, just append it as it is
                cleaned_column.append(evaluated_item)
        except (ValueError, SyntaxError):
            # In case of evaluation error or non-list contents, append the item as it is
            cleaned_column.append(item)
    return cleaned_column

# Clean the 'exports' and 'imports' columns
df['exports'] = clean_column_list(df['exports'].astype(str))
df['imports'] = clean_column_list(df['imports'].astype(str))
```

```python
import pandas as pd
import random

def calculate_and_distribute_demographics(df, region_info):
    # Function to distribute demographics to each settlement
    def distribute_demographics(row):
        region = row['region']
        settlement_population = row['settlement_population']
        breakdown = row['demographic_breakdown']

        # Skip if demographic_breakdown is already filled
        if breakdown:
            return breakdown

        # Initialize the breakdown if not present
        breakdown = {}
        region_totals = demographic_totals_by_region.get(region, {})

        # Randomize the order of demographics to be distributed
        demographics = list(region_totals.keys())
        random.shuffle(demographics)

        # Distribute demographics to the settlement based on the available totals
        for demo in demographics:
            total = region_totals[demo]
            if total > 0:
                # Allocate demographic based on what's needed vs. what's available
                to_distribute = min(total, settlement_population - sum(breakdown.values()))
                if to_distribute > 0:
                    breakdown[demo] = to_distribute
                    # Subtract the distributed number from the region's total
                    demographic_totals_by_region[region][demo] -= to_distribute

                # Print out the values subtracted and new totals
                print(f"Distributing {to_distribute} of {demo} to settlement {row['settlement']} in region {region}")
                print(f"New total for {demo} in region {region}: {demographic_totals_by_region[region][demo]}")

                # Break the loop if the settlement population is reached
                if sum(breakdown.values()) >= settlement_population:
                    break

        return breakdown

    # Apply the function to each row in the dataframe to distribute demographics
    df['demographic_breakdown'] = df.apply(distribute_demographics, axis=1)

    return df
```

```python
def calculate_demographic_distribution(total_population, demographic_percentages):
    return {demo: int(total_population * percent / 100) for demo, percent in demographic_percentages.items()}

def choose_primary_settlements(df_region, demographics_total):
    return {demo: random.choices(df_region.index, weights=df_region['settlement_population'], k=1)[0] for demo in demographics_total}

def distribute_demographics(df, df_region, demographics_total, primary_settlements):
    for demo, primary_settlement in primary_settlements.items():
        primary_allocation = min(demographics_total[demo], df.at[primary_settlement, 'settlement_population'])
        df.at[primary_settlement, 'demographic_breakdown'][demo] = primary_allocation
        demographics_total[demo] -= primary_allocation

        secondary_settlements = df_region.index[df_region.index != primary_settlement]
        for settlement in secondary_settlements:
            if demographics_total[demo] <= 0:
                break
            allocation = min(demographics_total[demo], df.at[settlement, 'settlement_population'])
            df.at[settlement, 'demographic_breakdown'][demo] = df.at[settlement, 'demographic_breakdown'].get(demo, 0) + allocation
            demographics_total[demo] -= allocation

def proportional_reduce(demographic_breakdown, population_limit):
    total_demographics = sum(demographic_breakdown.values())
    if total_demographics <= population_limit:
        return demographic_breakdown
    reduction_ratio = population_limit / total_demographics
    return {demo: int(count * reduction_ratio) for demo, count in demographic_breakdown.items()}

def redistribute_demographics(df, region_info):
    default_demo_breakdown = {'humans': 90, 'halflings': 9, 'other': 1}

    # Initialize 'demographic_breakdown' to empty dictionaries where it is NaN or not a dictionary
    df['demographic_breakdown'] = df['demographic_breakdown'].apply(
        lambda x: {} if pd.isna(x) or not isinstance(x, dict) else x
    )

    # Calculate total population per region
    region_totals = df.groupby('region')['settlement_population'].sum().to_dict()

    for region, info in region_info.items():
        # If region is not in DataFrame, continue to the next
        if region not in region_totals:
            continue

        # Get the DataFrame of just this region
        df_region = df[df['region'] == region].copy()

        # Only consider rows without a demographic breakdown
        df_region = df_region[df_region['demographic_breakdown'].apply(lambda x: not x)]

        # Calculate total demographic distribution for the region
        demographic_percentages = info.get('demographic_breakdown_percentages', default_demo_breakdown)
        demographics_total = calculate_demographic_distribution(region_totals[region], demographic_percentages)

        # Choose primary settlements for each demographic
        primary_settlements = choose_primary_settlements(df_region, demographics_total)

        # Distribute demographics across settlements
        distribute_demographics(df, df_region, demographics_total, primary_settlements)

        # Apply proportional reduction to settlements exceeding population limit
        for idx, row in df_region.iterrows():
            if not row['demographic_breakdown']:
                df.at[idx, 'demographic_breakdown'] = proportional_reduce(demographics_total, row['settlement_population'])

    return df

# Apply the function to the DataFrame
df = redistribute_demographics(df, region_info)
```
