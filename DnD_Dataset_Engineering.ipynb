```python
# Generating random data for each column

df['Region'] = np.random.choice(['Region1', 'Region2', 'Region3'], size=len(city_names))
df['Nation'] = np.random.choice(['Nation1', 'Nation2', 'Nation3'], size=len(city_names))
df['Population of City'] = np.random.randint(1000, 100000, size=len(city_names))
df['Government'] = np.random.choice(['Democracy', 'Monarchy', 'Oligarchy'], size=len(city_names))

# Generating random percentage data (between 0 and 100)
percentage_columns = [
    'Percentage of population Humans', 'Percentage of population Halflings',
    'Percentage of population Orcs', 'Percentage of population Elves',
    'Percentage of population Dwarves', 'Percentage of population Gnomes'
]
for col in percentage_columns:
    df[col] = np.random.uniform(0, 100, size=len(city_names))

df['Average Age of Population'] = np.random.uniform(20, 80, size=len(city_names))
df['Typical Cause of Death'] = np.random.choice(['Old Age', 'Disease', 'Accident'], size=len(city_names))
df['Wizarding School'] = np.random.choice(['Yes', 'No'], size=len(city_names))

# Generating random financial data (between 1000 and 1 million)
financial_columns = [
    'Active Economy (Gold)', 'Hidden Economy (Gold)',
    'Known Treasure Value', 'Rumored Treasure Value'
]
for col in financial_columns:
    df[col] = np.random.uniform(1000, 1000000, size=len(city_names))

df['Tax Rate'] = np.random.uniform(0, 50, size=len(city_names))

# Generating random list of strings for industries and resources
industries = ['Mining', 'Farming', 'Craftsmanship', 'Trading', 'Magic Goods']
resources = ['Gold', 'Silver', 'Wood', 'Stone', 'Magic Herbs']
df['Major Industries'] = [np.random.choice(industries, size=np.random.randint(1, len(industries)), replace=False).tolist() for _ in range(len(city_names))]
df['Major Resources'] = [np.random.choice(resources, size=np.random.randint(1, len(resources)), replace=False).tolist() for _ in range(len(city_names))]

# Generating random integer data for various columns
integer_columns = [
    '# of Landowners', '# Homeowners', 'Average Rent',
    '# of Peasants', 'Average Debt by Household',
    '# of Nobility', '# of Active Adventurers',
    'Number of Active Wizards', 'Dragon Sightings', 'Fate ranking'
]
for col in integer_columns:
    df[col] = np.random.randint(1, 1000, size=len(city_names))

# Generating random float data for climate and other factors
float_columns = [
    'Environmental Climate', 'Average Temperature',
    'Average Rainfall', 'Political Climate',
    'Magical Climate', 'Likelihood of Getting Cursed',
    'Danger Level', 'Openness to Strangers'
]
for col in float_columns:
    df[col] = np.random.uniform(0, 100, size=len(city_names))

# Generating random strings for landmarks
df['Notable Landmarks'] = np.random.choice(['Landmark1', 'Landmark2', 'Landmark3'], size=len(city_names))

# Save the updated DataFrame back to the Excel file
df.to_excel('/content/drive/MyDrive/Colab Notebooks/DnD/fictional_data.xlsx', index=False)
```
