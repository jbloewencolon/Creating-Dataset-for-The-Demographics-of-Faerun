```python
# Initialize the likelihood dictionary
race_region_likelihood = {region: {"Dwarf": 0, "Elf": 0, "Gnome": 0, "Half-Elf": 0, "Halfling": 0, "Human": 0} for region in regions}

# Increment the counts based on the regions lists
for region in dwarf_regions:
    race_region_likelihood[region]["Dwarf"] += 1

for region in elf_regions:
    race_region_likelihood[region]["Elf"] += 1

for region in gnome_regions:
    race_region_likelihood[region]["Gnome"] += 1

for region in half_elf_regions:
    race_region_likelihood[region]["Half-Elf"] += 1

for region in halfling_regions:
    race_region_likelihood[region]["Halfling"] += 1

for region in human_regions:
    race_region_likelihood[region]["Human"] += 1

# Calculate the likelihood
total_regions = len(regions)
for region in regions:
    for race in race_region_likelihood[region]:
        race_region_likelihood[region][race] /= total_regions
```

```python
def make_column_names_unique(df):
    seen = set()
    unique_columns = []
    for col in df.columns:
        new_col = col
        i = 1
        while new_col in seen:
            new_col = f"{col}_{i}"
            i += 1
        seen.add(new_col)
        unique_columns.append(new_col)
    df.columns = unique_columns
    return df

# Applying the function to make column names unique
df = make_column_names_unique(df)

# Now proceed with your existing script
from pandas import json_normalize

# First, you might need to convert the data to a list of dictionaries
data_dict_list = df.to_dict('records')

# Next, use json_normalize to flatten the dictionaries
flat_data_df = json_normalize(data_dict_list)

# Save the flat data DataFrame to an Excel file
flat_data_df.to_excel('/content/drive/MyDrive/Colab Notebooks/DnD/fictional_data.xlsx', index=False)
```

```python
def expand_column_data(cell):
    if isinstance(cell, dict):
        return pd.json_normalize(cell)
    elif isinstance(cell, list):
        # Check if the list contains dictionaries and if so, normalize them
        if cell and isinstance(cell[0], dict):
            return pd.json_normalize(cell)
        else:
            return pd.DataFrame(cell)
    else:
        # Assuming this represents a missing or other singular value
        return pd.DataFrame([cell])

# Apply the function to the 'Areas' column and combine the results into a new DataFrame
areas_data = df['Areas'].apply(expand_column_data)

# Now we have a series of DataFrames, we can combine them into a single DataFrame
areas_data_combined = pd.concat(areas_data.tolist(), ignore_index=True)

# Expanding other columns
geography_data = df['Geography'].apply(pd.Series)
government_data = df['Government'].apply(pd.Series)
demographics_data = df['Demographics'].apply(pd.Series)

# Concatenating expanded data with the original data frame
df = pd.concat([df, geography_data, government_data, areas_data_combined, demographics_data], axis=1)

# Dropping the original columns to avoid duplication
df = df.drop(columns=['Geography', 'Government', 'Areas', 'Demographics'])

df.head(40)
```
